```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
## alpha is to control the intensity of teh graph colour
## marker is used to change teh symbols/shape of the position wrt x and y axis
## grid to place a grid
## figure to fix the graph size

plt.figure(figsize=(6,2))
plt.scatter(x,y, c='#76b5c5', alpha=.9, marker='X')
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.title("This is x vs y graph")
plt.grid()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,2))
plt.bar(x,y, color='#b97455')
plt.xlabel("This is categotical data at x axis")
plt.ylabel("This is numerical data at y axis")
plt.title("This is a kind of bar grapth")
plt.grid()
```

```python
## to flip x and y , use barh (bar horizontal)

plt.figure(figsize=(6,2))
plt.barh(x,y, color='#b97455')
plt.xlabel("This is categotical data at x axis")
plt.ylabel("This is numerical data at y axis")
plt.title("This is a kind of bar grapth")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,9]
```

```python
## x axis should show the data
## y value should show the frequency

plt.figure(figsize = (5,2))
plt.hist(data, color = '#e07b39')
plt.xlabel("Representing unique value")
plt.ylabel("Representing frequency")
plt.title("Histogram Plot Example")
plt.grid()
plt.savefig('test.png')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z, c='r', alpha = .8)
plt.xlabel("This is x-axis")
plt.ylabel("This is y-axis")
plt.show()
```
