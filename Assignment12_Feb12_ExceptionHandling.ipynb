```python
''' Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
-------
Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence of a list/dictionary is invalid or does not exists. 
The two types of exceptions raised are: IndexError. KeyError.
 list that does not exist 
IndexError --- When we are trying to access an index (sequence) of ain that list or is out of range of that list, an index error is raised. '''

try:  
    a = ['a', 'b', 'c']  
    print (a[4])  
except LookupError:  
    print ("Index Error Exception Raised, list index out of range")
else:  
    print ("Success, no error!")

''' KeyError --- If a key you are trying to access is not found in the dictionary, a key error exception is raised. '''
try:  
    a = {1:'a', 2:'b', 3:'c'}  
    print (a[4])  
except LookupError:  
    print ("Key Error Exception Raised.")
else:  
    print ("Success, no error!")

```
