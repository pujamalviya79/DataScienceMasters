```python
'''Q3. Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point.
----------'''
from scipy.stats import norm
import numpy as np
  
data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points)
  
mean = np.mean(data)
std = np.std(data)
  
probability_pdf = norm.pdf(3, loc=mean, scale=std)
print(probability_pdf)
```

```python
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial 
distribution can be applied.
---------
Properties of Binomial Distribution are:
1: The number of observations n is fixed. 
2: Each observation is independent. 
3: Each observation represents one of two outcomes ("success" or "failure"). 
4: The probability of "success" p is the same for each outcome.

Examples:
If a coin is tossed 5 times, find the probability of:

1.Exactly 2 heads
Solution:
The repeated tossing of the coin is an example of a Bernoulli trial. According to the problem:
Number of trials: n=5
Probability of head: p= 1/2 and hence the probability of tail, q =1/2
For exactly two heads:
x=2
P(x=2) = 5C2 p2 q5-2 = 5! / 2! 3! × (½)2× (½)3
P(x=2) = 5/16

2.At least 4 heads.
Solution:
For at least four heads,
x ≥ 4, P(x ≥ 4) = P(x = 4) + P(x=5)
Hence,
P(x = 4) = 5C4 p4 q5-4 = 5!/4! 1! × (½)4× (½)1 = 5/32
P(x = 5) = 5C5 p5 q5-5 = (½)5 = 1/32
Therefore,
P(x ≥ 4) = 5/32 + 1/32 = 6/32 = 3/16
```

```python
Q7. How Binomial distribution different from Poisson distribution?
------------
Binomial distribution:
    Binomial distribution is one in which the probability of repeated number of trials are studied.	
    Biparametric nature
    Fixed number of trials
    Constant probability
    Only two possible outcomes, i.e. success or failure.
    Mean > Variance
    Example	: Coin tossing experiment.
Poisson Distribution:   
    Poisson Distribution gives the count of independent events occur randomly with a given period of time.
    Uniparametric nature
	Infinite number of trials
    Infinitesimal chance of success
    Unlimited number of possible outcomes.
    Mean = Variance
    Example : Printing mistakes/page of a large book.
```

```python
'''Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the 
sample mean and variance.
----------'''

import numpy as np
s = np.random.poisson(5, 1000)
mean = s.mean()
print("Mean :: ",mean)
variance = s.var()
print("Variance :: ",variance)
```

```python
Q9. How mean and variance are related in Binomial distribution and Poisson distribution?
---------
Binomial Distribution:
    Mean > Variance
Poisson Distribution:
    Mean = Variance
```
