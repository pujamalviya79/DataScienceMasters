```python
Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.
--------
mean = 50
sample std = 5
alpha = 0.05
suppose n=30

df = n-1 = 30-1=29

t-value = 2.756

Confidence interval, SE = s/sqrt(n)
        = 5/sqrt(30)
        = 5/5.47
        = 0.914
SE * t-value = 0.914*2.756
        = 2.52
    
upper = 50+2.52
    = 52.52
lower = 50-2.52
     = 47.48
    
Confidence interval for 95% confidence is 52.52 to 47.48
```

```python
'''Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.
-------------
expected_outcome=[20,20,20,10,10,20]
so, Ho=the color of M&Ms are as stated above
Ha=the color of M&Ms are not as stated above
lets take a sample of 300'''

import scipy.stats as stat
import numpy as np

expected_outcome = [60,60,60,30,30,60]
observed_outcome = [55,62,63,28,32,60]
sum(expected_outcome),sum(observed_outcome)

chisquare_test_statistics,p_value = stat.chisquare(observed_outcome,expected_outcome)
print(chisquare_test_statistics),print(p_value)

signigficance=0.05
dof=len(observed_outcome)-1
critical_value=stat.chi2.ppf(1-signigficance,dof)

if chisquare_test_statistics>critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

```python
Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.
-----------
Z α/2 = 1.96
According to the table, for a 95% CI, Z α/2 = 1.96.

p-hat: 60/500 = 0.12
q-hat: 1 – 0.12 = 0.88

For the lower percentage 
p-hat - Z sqrt(p(1-p)/n)
= 0.12 - 1.96(sqrt(0.0002112))
= 0.12 - 1.96*0.0145
= 0.12 – 0.02842 
= 0.09158

For the upper percentage
p-hat + Z sqrt(p(1-p)/n)
= 0.12 + 0.02842 
= 0.14842

Interval value = 0.09158*500 = 45.79  to  0.14842*500 = 74.21
So Confidence Interval is 48 to 74
```

```python
Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.
-----------
mean = 72
sample std = 10
alpha = 0.05
n=30

df = n-1 = 30-1=29

t-value = 1.725

Confidence interval, SE = s/sqrt(n)
        = 10/sqrt(30)
        = 10/5.47
        = 1.83
SE * t-value = 1.83*1.725
        = 3.16
    
upper = 72+3.16
    = 75.16
lower = 72-3.16
     = 68.84
    
Confidence interval is 68.84 to 75.16
We reject the hypothesis that population mean is significantly different from 70.
```
