```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

------
Class - A class is a user-defined blueprint from which objects are created.
Classes provide a means of bundling data and functionality together.
Classes are created by keyword class

Object - An object is an instance of a class.
A class is like blueprint while an instance is a copy of the class with actual values.

An object consists of:
    State -- Attributes
    Behaviour -- Methods

Example -- 
-- create a class

class myclass:
    x=5
    
-- create an object and print the value of x
p1 = myclass()
print()p1.x

Output -- 5
```

```python
Q3. Explain why the __init__() function is used. Give a suitable example.
----
All classes have a function called __init__, which is always executed when the class is being initiated.
__init__ function is used to assign values to object properties, or other operations that are necessary to do when teh object is being created.

Example --

class student():
    def __init__(self, name, studentId, Marks):
        self.name = name
        self.studentId = studentId
        self.Marks = Marks
        
s1 = student("Ram", 45, 78.5)

print(s1.name)
print(s1.studentId)
print(s1.Marks)

Output --
Ram
45
78.5
```

```python
Q4. Why self is used in OOPs?
----
The self parameter is a reference to the current instance of the class, and is used to access variables that belongs to the class.
self represents the instance of the class.
By using "self" we can access the attributes and methods of the class.
It binds the attributes with the given arguments.
```
