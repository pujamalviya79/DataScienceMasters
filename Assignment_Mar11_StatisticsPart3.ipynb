```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
----------
Z-Test:
    Z-test implies a hypothesis test which ascertains if the means of two datasets are different from each other when variance is given.
    Z-Test is based on Normal distribution
    For Z-test, population variance is Known
    Z-test is for Large sample
    Example:
        A teacher claims that the mean score of students in his class is greater than 82 with a standard deviation of 20. If a sample of 81 students was selected with a mean score of 90 then check if there is enough evidence to support this claim at a 0.05 significance level.
        Solution: As the sample size is 81 and population standard deviation is known, this is an example of a right-tailed one-sample z test.
        H0: μ=82
        H1: μ>82
        From the z table the critical value at α = 1.645
        x = 90, 
        μ = 82, n = 81, 
        σ = 20
            
        z = (X−μ)/σ/√n
        z = 3.6

        As 3.6 > 1.645 thus, the null hypothesis is rejected and it is concluded that there is enough evidence to support the teacher's claim.
        Answer: Reject the null hypothesis
T-Test:
    T-test refers to a type of parametric test that is applied to identify, how the means of two sets of data differ from one another when variance is not given.
	T-test is based on Student-t distribution
    For T-test, population variance is Unknown
    T-test is for Small sample
    Example:
        Let us take the example of a classroom of students that appeared for a test recently. 
        A sample of 10 students was chosen from a total of 150 students. 
        Calculate the sample’s t-test score if the mean score of the entire class is 78 and the mean score of the sample is 74 with a standard deviation of 3.5.
        Also, comment on whether the sample statistics are significantly different from the population at a 99.5% confidence interval.
        Solution:
            t = ( x̄ – μ) / (s / √n)
            t = (74 – 78) / (3.5 / √10)
            t = -3.61
            Therefore, the sample’s absolute t-test value is 3.61, which is less than the critical value (3.69) at a 99.5% confidence interval with a degree of freedom of 9. 
            So, the hypothesis of the sample statistic is different than the population and can be rejected.
```

```python
Q4: Explain Bayes's theorem with an example.
-----------
Bayes's theorem helps in determining the probability of an event that is based on some event that has already occurred.
It is a method to determine the probability of an event based on the occurrences of prior events. 
It is used to calculate conditional probability.
P(A|B)=(P(B|A)P(A))/P(B)
    Here, P(A) = how likely A happens(Prior knowledge)
        P(B) = how likely B happens(Marginalization)
        P(A/B) = how likely A happens given that B has happened(Posterior)
        P(B/A) = how likely B happens given that A has happened(Likelihood)
Example:
    Amy has two bags. Bag I has 7 red and 2 blue balls and bag II has 5 red and 9 blue balls. Amy draws a ball at random and it turns out to be red. 
    Determine the probability that the ball was from the bag I using the Bayes theorem.
    Solution: Let X and Y be the events that the ball is from the bag I and bag II, respectively. Assume A to be the event of drawing a red ball.
        We know that the probability of choosing a bag for drawing a ball is 1/2, that is,
        P(X) = P(Y) = 1/2
        Since there are 7 red balls out of a total of 11 balls in the bag I, therefore, 
        P(drawing a red ball from the bag I) = P(A|X) = 7/11
        Similarly, P(drawing a red ball from bag II) = P(A|Y) = 5/14
        We need to determine the value of P(the ball drawn is from the bag I given that it is a red ball), that is, P(X|A). 
        P(X|A) = (P(A|X)P(X)) / (P(A|X)P(X)+P(A|Y)P(Y))
                = [((7/11)(1/2))/(7/11)(1/2)+(5/14)(1/2)]
                = 0.64
        Answer: Hence, the probability that the ball is drawn is from bag I is 0.64
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
----------
A confidence interval is the mean of your estimate plus and minus the variation in that estimate. 
This is the range of values you expect your estimate to fall between if you redo your test, within a certain level of confidence.
Example:
    A random sample of 30 apples was taken from a large population. 
    On measuring their diameter the mean diameter of the sample was 91 millimeters with a standard deviation of 8 mm. 
    Calculate the 85% confidence limits for the mean diameter of the whole population of apples. 
    Solution:
        For 85% confidence, Z = 1.440 We have: ¯x=91, s=8, Z=1.440 and n=30
        CI = ¯x±(Z)*s/√n
        Therefore, the 85% confidence limits are:
        91±(1.440)×8/√30 
        = 91±(1.440)×8/5.477
        = 91±2.1 
    Answer: The 85% confidence limits are =91±2.1
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
---------
A margin of error tells you how many percentage points your results will differ from the real population value.
The larger the sample size, the smaller the sample standard error and therefore the smaller the margin of error.
Example:
    25 students in their final year were selected at random from a high school for a survey. 
    Among the survey participants, it was found that the average GPA (Grade Point Average) was 2.9 and the standard deviation of GPA was 0.5. 
    What is the margin of error, assuming 95% confidence level? Give correct interpretation.
    Solution: 
        the sample statistic is the sample mean which is the average GPA = 2.9
        Since population standard deviation is not known and the sample size is small, use a t distribution.
        Degrees of freedom=n−1=25−1=24
        α=1−Confidence level=1−0.95=0.05
        critical probability P* = 1-α/2 = 1-0.05/2 = 0.975
        The critical t value for cumulative probability of 0.975 and 24 degrees of freedom is 2.064
        
        Sample standard error = s/sqrt(n) = 0.5/sqrt(25) = 0.5/5 = 0.1
        
        Margin of error = critical value * sample standard error
                = 2.064 * 0.1 = 0.2064
        Interpretation: For a 95% confidence level, the average GPA is going to be 0.2064 points above and below the sample average GPA of 2.9.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
---------
n=50
sample mean = 6
sample std = 2.5

Ho: mean = 6
Ha: mean != 6

alpha = 0.05

df = n-1 = 50-1=49

t-value = 2.680

Confidence interval, SE = s/sqrt(n)
        = 2.5/sqrt(50)
        = 2.5/7.07
        = 0.35
SE * t-value = 0.35*2.680
        = 0.938
    
upper = 6+0.938
    = 6.938
lower = 6-0.938
    = 5.062
    
Confidence interval for 95% confidence is 5.062 to 6.938
```
