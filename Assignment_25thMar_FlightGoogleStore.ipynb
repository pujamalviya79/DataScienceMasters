```python
'''Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset
to identify the peak travel season. What features would you analyze to identify the peak season, and how
would you present your findings to your boss?
-------'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_excel('flight_price.xlsx')
df.head()

## Feature Engineering
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df.head()

fig = plt.figure(figsize =(10, 5))

v1=sns.barplot(x='month', y='Price', data=df, estimator=sum)
v1.set_title('Price and Month Comparison')
v1.set_ylabel('Price')
v1.set_xlabel('Month')
plt.show()

As per the above graph, we can say that May month has the peak price for Airlines
```

```python
'''Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?
-----'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_excel('flight_price.xlsx')
df.head()

df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]

df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)

df.head()

fig = plt.figure(figsize =(10, 5))

v1=sns.barplot(x='month', y='Price', data=df, estimator=sum)
v1.set_title('Price and Month Comparison')
v1.set_ylabel('Price')
v1.set_xlabel('Month')
plt.show()

'''As per the above graph, we can say that May month has the peak price for Airlines'''

#### with respect to number of stops and price graph
df.head()

v2=sns.barplot(x='Total_Stops', y='Price', data=df)
v2.set_title('Number of Stops and Price')
v2.set_ylabel('Price')
v2.set_xlabel('Total Stops')
plt.show()

'''As per above graph, we can say that flights having more stops are costlier , means as per stops are increasing, price is also increasing'''

plt.figure(figsize=(5,5))
s=df['Destination'].value_counts(ascending=False).plot(kind='pie',autopct='%1.0f%%', shadow=False)
plt.show()
'''As per above graph, we can say that flights for Cochin is highest and its approx half of the total flights (42%) and Bangalore has 27% flights'''

plt.figure(figsize=(5,5))
s=df['Source'].value_counts(ascending=False).plot(kind='pie',autopct='%1.0f%%', shadow=False)
plt.show()

'''As per the above graph, we can say that the most of the flights are from Delhi Source (42%) and second with Kolkata 27%'''

plt.figure(figsize=(30,10))
plt.title('Count of flights with different airlines')
ax=sns.countplot(x='Airline', data=df)
plt.xlabel('Airline Price')
plt.ylabel('Count of flights')

for p in ax.patches:
    ax.annotate(int(p.get_height()),(p.get_x()+0.25,p.get_height()+1),va='bottom',color='red')
    
'''As per above graph, we can say count of jet Airways flight is more as compared to other Airlines'''

```

```python
Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.
---------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline

df=pd.read_csv('https://github.com/pujamalviya79/DataScienceMasters/blob/main/googleplaystore.csv')


df['Reviews'].astype(int)
df['Reviews'].str.isnumeric().sum()
df_copy=df.copy()
df_copy=df_copy.drop(df_copy.index[10472])
## Convert Review Datatype to int
df_copy['Reviews']=df_copy['Reviews'].astype(int)

chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
        
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')

df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year

df_copy.to_csv('google_cleaned.csv')

df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')

df_copy['Category'].value_counts().plot.pie(y=df_copy['Category'],figsize=(15,16),autopct='%1.1f')
```
